```python
from pathlib import Path 
import pandas as pd 
import sys
import os

project_root = Path.cwd().parent.parent.parent
sys.path.insert(0, str(project_root))
```

```python
base_path =  Path("../../../data/raw/wipo/ip_indicators").resolve()
staging_path = Path("../../../data/staging/wipo/ip_indicators").resolve()


base_path.mkdir(parents=True, exist_ok=True)
staging_path.mkdir(parents=True, exist_ok=True)

# Ensure pandas prints every single column
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print(base_path)
print(staging_path)
```

```python
# Mapping for the short codes
type_map = {
    "industrial": "ID",
    "patent": "PA",
    "trademark": "TM"
}
```

```python
files = [
    "industrial_1b_office- Applications via the Hague system_Totals by office and residency type_1980_2024.csv",
    "industrial_1a_office- Direct applications_Totals by office and residency type_1980_2024.csv",
    "industrial_2b_origin- Registrations for applications via the Hague system_Totals by origin and residency type_1980_2024.csv",
    "industrial_2a_office- Registrations for direct applications_Totals by office and residency type_1980_2024.csv",
    "industrial_2a_origin- Registrations for direct applications_Totals by origin and residency type_1980_2024.csv",
    "industrial_2b_office- Registrations for applications via the Hague system_Totals by office and residency type_1980_2024.csv",
    "patent_1b_origin- PCT national phase entries_Totals by origin and residency type_1980_2024.csv",
    "industrial_1b_origin- Applications via the Hague system_Totals by origin and residency type_1980_2024.csv",
    "patent_1b_office- PCT national phase entries_Totals by office and residency type_1980_2024.csv",
    "patent_1a_office- Direct applications_Totals by office and residency type_1980_2024.csv",
    "patent_1a_origin- Direct applications_Totals by origin and residency type_1980_2024.csv",
    "industrial_1a_origin- Direct applications_Totals by origin and residency type_1980_2024.csv",
    "patent_2b_origin- Grants for PCT national phase entries_Totals by origin and residency type_1980_2024.csv",
    "patent_2b_office- Grants for PCT national phase entries_Totals by office and residency type_1980_2024.csv",
    "patent_2a_origin- Grants for direct applications_Totals by origin and residency type_1980_2024.csv",
    "patent_2a_office- Grants for direct applications_Totals by office and residency type_1980_2024.csv",
    "trademark_1b_origin- Applications via the Madrid system_Totals by origin and residency type_1980_2024.csv",
    "trademark_1a_office- Direct applications_Totals by office and residency type_1980_2024.csv",
    "trademark_1a_origin- Direct applications_Totals by origin and residency type_1980_2024.csv",
    "trademark_1b_office- Applications via the Madrid system_Totals by office and residency type_1980_2024.csv",
    "trademark_2b_origin- Registrations for applications via the Madrid system_Totals by origin and residency type_1980_2024.csv",
    "trademark_2a_office- Registrations for direct applications_Totals by office and residency type_1980_2024.csv",
    "trademark_2a_origin- Registrations for direct applications_Totals by origin and residency type_1980_2024.csv",
    "trademark_2b_office- Registrations for applications via the Madrid system_Totals by office and residency type_1980_2024.csv"
]
```

```python
for filename in files:
    path = os.path.join(base_path, filename)
    if not os.path.exists(path): continue


    df = pd.read_csv(path, header=6)
    
    # Drop Code (index 1) and Missing (index 2)
    df.drop(df.columns[[1, 2]], axis=1, inplace=True)
    
    # Trim empty years from the right
    while len(df.columns) > 2:
        if df[df.columns[-1]].isna().all():
            df.drop(df.columns[-1], axis=1, inplace=True)
        else:
            break

    # Rename Headers
    num_year_cols = len(df.columns) - 2 
    years = [str(1980 + i) for i in range(num_year_cols)]
    
    if "office" in filename.lower():
        df.columns = ['Office', 'Type'] + years
    elif "origin" in filename.lower():
        df.columns = ['Origin', 'Type'] + years



    code_part = filename.split('-')[0].strip()
    
    #  Split into components (e.g., ['industrial', '1b', 'office'])
    parts = code_part.split('_')
    
    if len(parts) >= 3:
        category = parts[0] 
        code_num = parts[1]  
        location = parts[2]  
        
        #  Map category to short code (ID, PA, TM)
        short_cat = type_map.get(category, category)
        
        #  Construct new name
        new_name = f"stg_wipo__{short_cat}{code_num}_{location}.parquet"
        
        # export
        out_file = staging_path / new_name
        df.to_parquet(out_file, index=False)
        print(f"Exported: {new_name}")
    else:
        print(f"Skipping naming for: {filename}")
```
