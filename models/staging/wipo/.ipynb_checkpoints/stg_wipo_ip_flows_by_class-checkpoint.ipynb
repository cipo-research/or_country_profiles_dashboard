```python
from pathlib import Path
import pandas as pd
import sys
import os
```

```python
# PATH SETUP

project_root = Path.cwd().parent.parent.parent
sys.path.insert(0, str(project_root))


base_path = Path("../../../data/raw/wipo/ip_indicators").resolve()
staging_path = Path("../../../data/staging/wipo/ip_indicators").resolve()

base_path.mkdir(parents=True, exist_ok=True)
staging_path.mkdir(parents=True, exist_ok=True)

# Ensure pandas prints every single column
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print(f"Reading Raw from: {base_path}")
print(f"Saving Staging to: {staging_path}")
```

```python
# Mapping short codes (patent -> PA)
type_map = {
    "industrial": "ID",
    "patent": "PA",
    "trademark": "TM"
}
```

```python
# The specific files for "IP Flows by Class/Technology"
files = [
    # --- Patents (By Technology) ---
    "patent_4a - Patent publications by technology_Count by filling office and applicant's origin_1980_2024.csv",
    "patent_5 - Patent grants by technology_Count by filing office and applicant's origin_1980_2024.csv",

    # --- Trademarks (By Nice Class) ---
    "trademark_3a- Direct applications by class_Count by filing office and applicant's origin_1980_2024.csv",
    "trademark_3b- Applications via the Madrid system by class_Count by filing office and applicant's origin_1980_2024.csv",
    "trademark_4a- Registrations for direct applications by class_Count by filing office and applicant's origin_1980_2024.csv",
    "trademark_4b- Registrations for applications via the Madrid system by class_Count by filing office and applicant's origin_1980_2024.csv",

    # --- Industrial Designs (By Locarno Class) ---
    "industrial_3a- Direct applications by class_Count by filing office and applicant's origin_1980_2024.csv",
    "industrial_3b- Applications via the Hague system by class_Count by filing office and applicant's origin_1980_2024.csv",
    "industrial_4a- Registrations for direct applications by class_Count by filing office and applicant's origin_1980_2024.csv",
    "industrial_4b- Registrations for applications via the Hague system by class_Count by filing office and applicant's origin_1980_2024.csv"
]
```

```python
for filename in files:
    path = base_path / filename
    if not path.exists():
        print(f"[MISSING] {filename}")
        continue

    # Read Data (Skip WIPO metadata rows)
    try:
        df = pd.read_csv(path, header=6)
    except Exception as e:
        print(f"[ERROR] Could not read {filename}: {e}")
        continue

    # Extract Code (e.g., 'patent_4a')
    code_part = filename.split('-')[0].strip()
    
    # Clean Columns
    # These 'Class' files usually have 4 metadata columns: 
    # [Office, Office Code, Origin, CLASS_COLUMN, 1980, 1981...]
    
    # Check if we have enough columns to proceed
    if len(df.columns) < 5:
        print(f"[SKIP] {filename} - Not enough columns.")
        continue

    # Rename the first 4 columns standardly
    # The 4th column is dynamic (Field of technology / Nice class / Locarno class)
    class_col_original_name = df.columns[3]
    
    # Map to standardized names
    # Drop 'office_code' (col index 1) later
    df.columns.values[0] = "Office"
    df.columns.values[1] = "office_code" 
    df.columns.values[2] = "Origin"
    
    # Keep the class column name somewhat generic for now or map it
    if "technology" in str(class_col_original_name).lower():
        df.rename(columns={class_col_original_name: "tech_field"}, inplace=True)
    elif "nice" in str(class_col_original_name).lower():
        df.rename(columns={class_col_original_name: "nice_class"}, inplace=True)
    elif "locarno" in str(class_col_original_name).lower():
        df.rename(columns={class_col_original_name: "locarno_class"}, inplace=True)
    else:
        # Fallback if header is weird
        df.rename(columns={df.columns[3]: "class_code"}, inplace=True)

    # Drop the office code
    df.drop("office_code", axis=1, inplace=True)

    # Trim empty years from the right
    while df.columns[-1].isdigit() and df[df.columns[-1]].isna().all():
        df.drop(df.columns[-1], axis=1, inplace=True)

    #  Generate Output Filename (stg_wipo__PA4a.parquet)
    parts = code_part.split('_')
    if len(parts) >= 2:
        short_cat = type_map.get(parts[0], parts[0]) # patent -> PA
        indicator_code = parts[1] # 4a
        new_name = f"stg_wipo__{short_cat}{indicator_code}.parquet"
        
        output_file = staging_path / new_name
        df.to_parquet(output_file, index=False)
        print(f"Exported: {new_name}")
```
