## 2. Load & transform raw WIPO data

```python
from pathlib import Path 
import pandas as pd 

from utils.paths import RAW_IP_INDICATORS_DIR, STG_IP_INDICATORS_DIR
```

```python
# -------------------------------
# PARAMETERS FOR HELPER FUNCTIONS
# -------------------------------

# Staged column names for `unpivot_indicator_df()` and `rename_indicator_cols()`
DESTINATION_OFFICE      = 'destination_office'
DESTINATION_OFFICE_CODE = 'destination_office_st3_code'
ORIGIN      = 'origin'
ORIGIN_CODE = 'origin_st3_code'
FIELD_OF_TECHNOLOGY    = 'tech_field'
LOCARNO_CLASSIFICATION = 'locarno_class'
NICE_CLASSIFICATION    = 'nice_class'
YEAR  = 'year'      # Column header for `unpivot_indicator_df()`
COUNT = 'count'     # Column header for `unpivot_indicator_df()`

COL_RENAME_MAP = {  # Column map for `rename_indicator_cols()`
    'Office':        DESTINATION_OFFICE,
    'Office (Code)': DESTINATION_OFFICE_CODE,
    'Origin':        ORIGIN,
    'Origin (Code)': ORIGIN_CODE,
    'Field of technology':    FIELD_OF_TECHNOLOGY,     # Patents
    'Locarno classification': LOCARNO_CLASSIFICATION,  # Industrial designs
    'Nice classification':    NICE_CLASSIFICATION,     # Trademarks
}

# Staged column dtypes for `enforce_indicator_schema()`
NUMERIC_COLS = [YEAR, COUNT]
STRING_COLS = [
    DESTINATION_OFFICE,
    DESTINATION_OFFICE_CODE,
    ORIGIN,
    ORIGIN_CODE,
    FIELD_OF_TECHNOLOGY,
    LOCARNO_CLASSIFICATION,
    NICE_CLASSIFICATION,
]


# ----------------
# HELPER FUNCTIONS
# ----------------

# Order of operations for raw -> staging transformations:
#   1. load_indicator_df()
#   2. unpivot_indicator_df()
#   3. rename_indicator_cols()
#   4. enforce_indicator_schema()

def load_indicator_df(filepath: Path, verbose: bool = False) -> pd.DataFrame:
    if verbose == True:
        print(f"Loading IP indicator '{filepath.name}'")
    return pd.read_csv(filepath, skiprows=6, index_col=False)


def unpivot_indicator_df(df: pd.DataFrame) -> pd.DataFrame:
    """
    Unpivot wide WIPO year data. Individiual year columns moved to a
    'Year' column; data for each year moved to a 'Count' column.

    Input pd.DataFrame
    ------------------
        | Office | ... | Origin | 1985 | 1986 | ... | 2023 |
        |--------|-----|--------|------|------|-----|------|
        | Canada | ... | Brazil |   10 |   15 | ... |  100 |
            ⋮       ⋮      ⋮        ⋮      ⋮     ⋮      ⋮

    Returns
    -------
        | Office | ... | Origin | year | count |
        |--------|-----|--------|------|-------|
        | Canada | ... | Brazil | 1985 |    10 |
        | Canada | ... | Brazil | 1986 |    15 |
            ⋮       ⋮      ⋮       ⋮        ⋮
        | Canada | ... | Brazil | 2023 |   100 |
            ⋮       ⋮      ⋮       ⋮        ⋮
    """
    df.columns = [str(col).strip() for col in df.columns]
    cols = [col for col in df.columns]
    
    first_year_col_idx = next(i for i, col in enumerate(cols) if col.isdigit())
    id_vars = cols[:first_year_col_idx]
    year_cols = cols[first_year_col_idx:]

    # Un-pivot year cols to the column 'year'; add new col 'count'
    return df.melt( id_vars=id_vars, value_vars=year_cols,
                   var_name=YEAR,    value_name=COUNT)


def rename_indicator_cols(df: pd.DataFrame) -> pd.DataFrame:
    """
    Run unpivot_indicator_df() BEFORE this.
    """
    cols_to_rename = {col: new_col
                      for col, new_col in COL_RENAME_MAP.items()
                      if col in df.columns}

    return df.rename(columns=cols_to_rename)


def enforce_indicator_schema(
    df: pd.DataFrame,
    strict:  bool = False,
    verbose: bool = False  # For column header & dtype debugging
    ) -> pd.DataFrame:
    """
    Run rename_indicator_cols() BEFORE this.

    Enforce staged schema:
      - STRING_COLS  -> Pandas `string` with trimming and '' set to <NA>
      - NUMERIC_COLS -> Pandas-assigned numeric dtype

    If  strict=True, then throw error when required columns are missing.
    If verbose=True, then print headers of string/numeric cols being cleaned.
    """
    def clean_string_col(s: pd.Series) -> pd.Series:
        s = s.astype('string')
        s = s.str.strip()
        return s.replace('', pd.NA)

    missing = [col for col in (STRING_COLS + list(NUMERIC_COLS)) if col not in df.columns]
    if strict and missing:
        raise ValueError(f"Missing required columns: {missing}")
    
    for col in STRING_COLS:
        if col in df.columns:
            df[col] = clean_string_col(df[col])
            if verbose:
                print(f"Cleaned string col: '{df[col].name}'")

    for col in NUMERIC_COLS:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            if verbose:    
                print(f"  Cast numeric col: '{df[col].name}' -> {df[col].dtype}")

    return df
```

### 2.1 Map filenames

```python
# WIPO IP indicator mappings
ip_indicator_map = {
    'patent':     ('PA', ['1a', '1b',
                          '2a', '2b',
                          '3',  '5',  '9',  '10']),
    'trademark':  ('TM', ['1a', '1b',
                          '2a', '2b',
                          '4a', '4b',
                          '5a', '5b',
                          '6a', '6b',
                          '7',  '8',  '9']),
    'industrial': ('ID', ['1a', '1b',
                          '2a', '2b',
                          '4a', '4b',
                          '5a', '5b', 
                          '6a', '6b',
                          '7',  '8',  '9']),
}

# Create map: { 'industrial_1a': 'ID1a', ... }
ind_name_to_abbr: dict[str, str] = {
    f'{ip_type}_{code}': f'{abbr}{code}'
    for ip_type, (abbr, codes) in ip_indicator_map.items()
    for code in codes
}

# Create map: { 'industrial_1a': <Path('.../stg_wipo__ID1a.parquet')>, ... }
raw_to_stg: dict[str, Path] = {
    prefix: (STG_IP_INDICATORS_DIR / f'stg_wipo__{abbr}.parquet')
    for prefix, abbr in ind_name_to_abbr.items()
}
```

### 2.2 Find raw WIPO IP indicator files

```python
# Check that output directory exists
STG_IP_INDICATORS_DIR.mkdir(parents=True, exist_ok=True)

# Create map of raw file matches
matches_map: dict[str, list[Path]] = {
    prefix: sorted(RAW_IP_INDICATORS_DIR.glob(f'{prefix}*.csv'))
    for prefix in raw_to_stg.keys()
}
```

### 2.3 Validate that we have all raw WIPO IP indicator files

```python
# Find missing and duplicate files; raise an error if so
missing_raw_files: list[str] = sorted(
    [prefix for prefix, files in matches_map.items() if not files]
)
duplicate_raw_files: dict[str, list[Path]] = {
    prefix: files for prefix, files in matches_map.items()
    if len(files) > 1
}

if missing_raw_files or duplicate_raw_files:
    parts = []
    if missing_raw_files:
        parts.append('Missing prefixes: ' + ', '.join(missing_raw_files))
    if duplicate_raw_files:
        duplicates: str = '; '.join(f"{prefix}: [{', '.join(f.name for f in files)}]"
                                    for prefix, files in duplicate_raw_files.items())
        parts.append('Duplicate matches per prefix: ' + duplicates)
    raise ValueError('Input validation failed: ' + ' | '.join(parts))
```

### 2.4 Process raw data and save in staging layer

```python
# For each raw file: Load CSV -> Unpivot -> Save as Parquet
prefix_padding   = max((len(key) for key in matches_map.keys()))  # Finds longest str in ('patent_1a', ..., 'industrial_9')
out_path_padding = max((len(filepath.name) for filepath in raw_to_stg.values()))  # Finds longest filename in ('stg_wipo__PA1a.parquet', ...)

for prefix, files in matches_map.items():
    raw_file = files[0]  # Ensure only 1 file is selected per IP indicator
    out_path = raw_to_stg[prefix]

    raw_df = load_indicator_df(raw_file)
    stg_df = unpivot_indicator_df(raw_df)
    stg_df = rename_indicator_cols(stg_df)
    stg_df = enforce_indicator_schema(stg_df)

    print(f'Writing DataFrame: {prefix:<{prefix_padding}} '
          f'to {out_path.name:.<{out_path_padding}}... ', end='')
    stg_df.to_parquet(out_path, compression='snappy', index=False)
    print('Done')
```
