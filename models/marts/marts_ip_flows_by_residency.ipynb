```python
import pandas as pd
import os, re, sys
from pathlib import Path
import duckdb

# Add project root to path FIRST (before any imports from utils)
project_root = Path.cwd().parent.parent
sys.path.insert(0, str(project_root))

# NOW import from utils
from utils.paths import (INT_IP_FLOWS_DIR,
                         MARTS_IP_FLOWS_DIR)
```

```python
int_filename  = 'int_ip_flows_by_residency.parquet'
mart_filename = 'fct_ip_flows_by_residency.csv.gz'

in_path  = INT_IP_FLOWS_DIR  / int_filename
out_path = MARTS_IP_FLOWS_DIR / mart_filename
```

```python
con = duckdb.connect()

query = f"""
COPY (SELECT * FROM read_parquet('{in_path.as_posix()}'))
TO '{out_path.as_posix()}'
(FORMAT CSV,
 HEADER,
 DELIMITER ',',
 QUOTE '"',
 ESCAPE '"',
 NULL '',
 COMPRESSION GZIP);
"""

con.execute(query)

con.close()
```
