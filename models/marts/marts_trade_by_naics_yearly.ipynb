```python
# Mart: Final fact table

from time import sleep
from pathlib import Path
import sys
import duckdb as dd

project_root = Path.cwd().parent.parent
sys.path.insert(0, str(project_root))

from utils.paths import (
    INT_MACROECON_FLOWS_DIR,
    MARTS_MACROECON_FLOWS_DIR,
)

IN_TBL_NAME = "int_trade_by_naics"
IN_FILEPATH = INT_MACROECON_FLOWS_DIR / f"{IN_TBL_NAME}.parquet"

OUT_TBL_NAME = "fct_trade_by_naics_yearly"
OUT_FILEPATH = MARTS_MACROECON_FLOWS_DIR / f"{OUT_TBL_NAME}.parquet"

con = dd.connect()
```

```python
mart_query = f"""
-- Final fact table 


SELECT
    year,
    trade_type,
    trading_partner,
    naics_code,
    naics_name,
    cad_value
FROM
    read_parquet('{IN_FILEPATH}')
ORDER BY
    year,
    trading_partner,
    trade_type,
    naics_name
"""

con.sql(f"CREATE OR REPLACE VIEW '{OUT_TBL_NAME}' AS ({mart_query})")
```

```python
save_query = f"""
COPY
    (SELECT * FROM '{OUT_TBL_NAME}')
TO
    '{OUT_FILEPATH.as_posix()}'
(FORMAT PARQUET);
"""

con.sql(save_query)
con.close()
```
