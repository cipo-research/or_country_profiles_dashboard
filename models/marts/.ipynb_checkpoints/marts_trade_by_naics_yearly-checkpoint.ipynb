Mart - Ranked annual trade by NAICS class
=========================================

```python
from time import sleep
from pathlib import Path 
import pandas as pd 
import sys

project_root = Path.cwd().parent.parent
sys.path.insert(0, str(project_root))
import duckdb as dd

from utils.paths import (INT_MACROECON_FLOWS_DIR,
                         MARTS_MACROECON_FLOWS_DIR)
```

```python
# Notebook params
IN_TBL_NAME = 'int_trade_by_naics'
IN_FILEPATH = INT_MACROECON_FLOWS_DIR / f'{IN_TBL_NAME}.parquet'

OUT_TBL_NAME = 'fct_trade_by_naics_yearly'
OUT_FILEPATH = MARTS_MACROECON_FLOWS_DIR / f'{OUT_TBL_NAME}.parquet'
```

```python
con = dd.connect()
```

```python
mart_query = f"""
-- This is the final mart model for consumption.
-- 1. Calculates the dense rank of each NAICS industry's trade value.
-- 2. The rank is partitioned by year, trade type, AND trading partner.
-- 3. Sets the rank to NULL for the 'Total' industry row as it shouldn't be ranked.

WITH aggregated_data AS (
    SELECT
        *
    FROM
        read_parquet('{IN_FILEPATH}')
),
ranked_data AS (
    SELECT
        *,
        DENSE_RANK() OVER (
            PARTITION BY year, trade_type, trading_partner
            ORDER BY cad_value DESC NULLS LAST
        ) AS rank_yearly
    FROM
        aggregated_data
)
SELECT
    year,
    trading_partner,
    trade_type,
    trade_direction,
    naics_industry_grp,
    naics_name,
    naics_code,
    cad_value,
    -- Apply the business rule: do not rank the 'Total' row
    CASE
        WHEN naics_code IS NOT NULL THEN rank_yearly
        ELSE NULL
    END AS rank_yearly
FROM
    ranked_data
ORDER BY
    year, trading_partner, trade_type, rank_yearly NULLS LAST
"""

con.sql(f"CREATE OR REPLACE VIEW '{OUT_TBL_NAME}' AS ({mart_query})")
```

```python
save_query = f"""
COPY
    (SELECT * FROM '{OUT_TBL_NAME}')
TO
    '{OUT_FILEPATH.as_posix()}'
(FORMAT parquet);
"""

con.sql(save_query)
con.close()
```
