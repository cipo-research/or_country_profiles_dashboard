```python

import duckdb

from utils.paths import (INT_IP_FLOWS_DIR,
                         MARTS_IP_FLOWS_DIR)
```

```python
int_filename  = 'int_ip_flows_by_class.parquet'
mart_filename = 'fct_ip_flows_by_class_yearly.csv.gz'

in_path  = INT_IP_FLOWS_DIR  / int_filename
out_path = MARTS_IP_FLOWS_DIR / mart_filename
```

```python
con = duckdb.connect()

query = f"""
COPY (SELECT * FROM read_parquet('{in_path.as_posix()}'))
TO '{out_path.as_posix()}'
(FORMAT CSV,
 HEADER,
 DELIMITER ',',
 QUOTE '"',
 ESCAPE '"',
 NULL '',
 COMPRESSION GZIP);
"""

con.execute(query)

con.close()
```
