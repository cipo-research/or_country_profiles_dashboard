```python
import os, re, sys
from pathlib import Path
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

project_root = Path.cwd().parent.parent
sys.path.insert(0, str(project_root))
```

```python
import duckdb

from utils.paths import INT_IP_FLOWS_DIR, MARTS_IP_FLOWS_DIR
```

```python
int_filename   = 'int_ip_flows.parquet'
marts_filename = 'fct_ip_flows_yearly.csv.gz'

in_path  = INT_IP_FLOWS_DIR   / int_filename
out_path = MARTS_IP_FLOWS_DIR / marts_filename
```

```python
con = duckdb.connect()

query = f"""
COPY (SELECT * FROM read_parquet('{in_path.as_posix()}'))
TO '{out_path.as_posix()}'
(FORMAT CSV,
 HEADER,
 DELIMITER ',',
 QUOTE '"',
 ESCAPE '"',
 NULL '',
 COMPRESSION GZIP);
"""

con.execute(query)

con.close()
```
