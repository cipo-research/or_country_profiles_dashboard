Intermediate: Macroeconomic flows to and from Canada
====================================================

Flow types included:
- Imports
- Exports
- Foreign direct investment by immediate investor
- Foreign direct investment by ultimate investor

```python
from time import sleep
from pathlib import Path 
import pandas as pd 
import sys

project_root = Path.cwd().parent.parent
sys.path.insert(0, str(project_root))

from utils.paths import (STG_STATCAN_DIR, STG_GAC_DIR, STG_UN_DIR,
                         INT_MACROECON_FLOWS_DIR)
from utils.etl_utils import merge_columns_safe
```

```python
# Load the staged economic data
stg_imports_df = pd.read_csv(STG_STATCAN_DIR / 'stg_statcan__imports.csv')
stg_exports_df = pd.read_csv(STG_STATCAN_DIR / 'stg_statcan__exports.csv')
stg_fdi_immediate_df = pd.read_csv(STG_STATCAN_DIR / 'stg_statcan__fdi_immediate.csv')
stg_fdi_ultimate_df  = pd.read_csv(STG_STATCAN_DIR / 'stg_statcan__fdi_ultimate.csv')

# Load the staged region dimension data
gac_df   = pd.read_csv(STG_GAC_DIR / 'stg_gac__regions.csv')
un_df    = pd.read_csv(STG_UN_DIR  / 'stg_un__m49_regions.csv')
sccai_df = pd.read_csv(STG_STATCAN_DIR / 'stg_statcan__sccai_regions.csv')
```

```python
m49_to_sccai = (
    sccai_df[['m49_code', 'sccai_code']]
    .dropna(subset=['m49_code'])
    .drop_duplicates(subset=['m49_code'])
)

region_to_sccai = (
    sccai_df[['region', 'sccai_code']]
    .dropna(subset=['region'])
    .drop_duplicates(subset=['region'])
)
```

```python
def add_region_codes(df: pd.DataFrame, region_col_name: str) -> pd.DataFrame:
    """
    Joins a DataFrame with various region dimension tables to add
    UN M49 country codes and SCCAI country codes.
    """ 
    # Join df with GAC and UN M49 data
    merged_gac_un = (
        df
        .merge(gac_df[['gc_name_abbr_en', 'm49_code']],
               left_on=region_col_name,
               right_on='gc_name_abbr_en',
               how='left')
        .merge(un_df[['m49_name', 'm49_code']],
               left_on=region_col_name,
               right_on='m49_name',
               how='left',
               suffixes=('_from_gac', '_from_un'))
    )

    # Coalesce M49 codes from UN and GAC sources
    merged_gac_un['m49_code'] = merge_columns_safe(df=merged_gac_un,
                                                     col1='m49_code_from_gac',
                                                     col2='m49_code_from_un')
    
    # 2. Merge df+M49 with SCCAI regions
    merged_gac_un_sccai = (
        merged_gac_un
        .merge(m49_to_sccai,
               on='m49_code',
               how='left')
        .merge(region_to_sccai,
               left_on=region_col_name,
               right_on='region',
               how='left',
               suffixes=('_from_m49', '_from_region'))
    )
    
    # Coalesce SCCAI codes
    merged_gac_un_sccai['sccai_code'] = merge_columns_safe(df=merged_gac_un_sccai,
                                                            col1='sccai_code_from_m49',
                                                            col2='sccai_code_from_region')
    
    # Assign 'All countries' the M49 code for 'World'
    merged_gac_un_sccai.loc[merged_gac_un_sccai[region_col_name] == 'All countries', 'm49_code'] = 1
    
    final_cols = list(df.columns) + ['m49_code', 'sccai_code']
    
    return merged_gac_un_sccai[final_cols].convert_dtypes()
```

```python
imports_df = add_region_codes(stg_imports_df, 'imports_from_region')
exports_df = add_region_codes(stg_exports_df, 'exports_to_region')
fdi_immediate_df = add_region_codes(stg_fdi_immediate_df, 'region')
fdi_ultimate_df  = add_region_codes(stg_fdi_ultimate_df,  'region')

imports_final = (
    imports_df
    .assign(flow_type='Imports from region')
    .rename(columns={'imports_from_region': 'region'})
)
exports_final = (
    exports_df
    .assign(flow_type='Exports to region')
    .rename(columns={'exports_to_region': 'region'})
)

fdi_type_map = {
    # FDI by immediate investor
    'Canadian direct investment abroad - total book value':   'FDI immediate outward, total book value',
    'Foreign direct investment in Canada - total book value': 'FDI immediate inward, total book value',
    
    # FDI by ultimate investor
    'Canadian direct investment abroad by ultimate investor country – total book value':   'FDI ultimate outward, total book value',
    'Foreign direct investment in Canada by ultimate investor country - total book value': 'FDI ultimate inward, total book value',
}
fdi_immediate_final = fdi_immediate_df.assign(
    flow_type=fdi_immediate_df['fdi_type'].map(fdi_type_map)).drop(columns=['fdi_type']
)
    
fdi_ultimate_final = fdi_ultimate_df.assign(
    flow_type=fdi_ultimate_df['fdi_type'].map(fdi_type_map)).drop(columns=['fdi_type']
)
```

```python
# Concatenate economic flow dfs
int_macroecon_flows = pd.concat([imports_final, exports_final,
                                 fdi_immediate_final, fdi_ultimate_final],
                                ignore_index=True)

col_order = ['year', 'region', 'm49_code', 'sccai_code', 'flow_type', 'dollar_value']

int_macroecon_flows = (
    int_macroecon_flows[col_order]
    .drop_duplicates(ignore_index=True)
    .convert_dtypes()
)
```
