Intermediate IP flows by class
==============================

Using WIPO IP indicators: PA5, TM4a/b, ID4a/b

| **IP Type** | **Code** | **Columns** | **Description** |
|---|---|---|---|
| **Patents** | PA5  | Office, Office (Code), Origin, Field of technology, year, count | Patent grants by technology |
| **Trademarks** | TM4a | Office, Office (Code), Origin, Nice classification, year, count | Registrations for direct applications by class |
| **Trademarks** | TM4b | Office, Office (Code), Origin, Nice classification, year, count | Registrations for applications via the Madrid system by class |
| **Industrial designs** | ID4a | Office, Office (Code), Origin, Locarno classification, year, count | Registrations for direct applications by class |
| **Industrial designs** | ID4b | Office, Office (Code), Origin, Locarno classification, year, count | Registrations for applications via the Hague system by class |

```python
import pandas as pd

from utils.paths import STG_IP_INDICATORS_DIR, INT_IP_FLOWS_DIR

```

```python
# Load staged IP indicators into a dict of dataframes

IP_FLOWS_BY_CLASS_INDICATORS = ['PA5', 'ID4a', 'ID4b', 'TM4a', 'TM4b']

dfs: dict[str, pd.DataFrame] = {
    code: pd.read_parquet(STG_IP_INDICATORS_DIR / f'stg_wipo__{code}.parquet')
    for code in IP_FLOWS_BY_CLASS_INDICATORS
}
```

```python
for ip_indicator, df in dfs.items():
    print(f'{ip_indicator} columns:')
    print(list(dfs[ip_indicator].columns)) 
    print()
    
```

```python
for ip_indicator, df in dfs.items():
    print(f'{ip_indicator} col dtypes:')
    for col in df.columns:
        print(f'   {col}: {df[col].dtype}') 
    print()
    
```

```python
classification_map = {
    'tech_field':    'Tech field',  # Patents
    'locarno_class': 'Locarno',     # Industrial designs
    'nice_class':    'Nice',        # Trademarks
}
for ip_indicator in dfs:
    df = dfs[ip_indicator]
    df['ip_indicator'] = ip_indicator

    # Find which IP classification column is present in each df
    found = None
    for col in classification_map:
        if col in df.columns:
            found = col
            break
    
    # Assign a new column 'class_type' that represents the classification
    # then rename the IP classification column
    if found:        
        df['class_type'] = classification_map[found]
        
        # Rename IP classification header col to 'class_code'
        # Note: This is faster than df.rename()
        cols = df.columns.to_list()
        found_idx = cols.index(found)
        cols[found_idx] = 'class_code'
        df.columns = cols
```

```python
dfs['ID4a']
```

```python

reference_df = next(iter(dfs.values()))     # Get first dataframe in `dfs`
reference_cols = set(reference_df.columns)  # Get cols from first dataframe in `dfs`

# Check whether columns of all dataframes match so we can concatenate them
all_match = all(set(df.columns) == reference_cols for df in dfs.values())

all_classes_df = pd.DataFrame()
if all_match:
    print('All cols match. Concatenating... ', end='')
    # Concat along rows
    all_classes_df = (
        pd.concat(dfs.values(), axis=0, ignore_index=True)
    )
    print('Done.')
else:
    print('Not all cols match. Abort concatenation.')
```

```python
# Order cols and cast cols to correct datatypes

col_order = [
    'year',
    'origin',
    'destination_office',
    'destination_office_st3_code',
    'ip_indicator',
    'class_type',
    'class_code',
    'count'
]

all_classes_df = all_classes_df[col_order].convert_dtypes()
```

```python
all_classes_df

```

```python
all_classes_filename = 'int_ip_flows_by_class.parquet'

all_classes_df.to_parquet(INT_IP_FLOWS_DIR / all_classes_filename, compression='snappy', index=False)
```
