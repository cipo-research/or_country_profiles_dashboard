## 0. Initial setup

```python
# TODO: This ETL notebook is a WIP; finish!

import pandas as pd

from utils.paths import (STG_IP_INDICATORS_DIR, STG_CHINA_DIR, STG_IP_OFFICES_DIR,
                         INT_IP_FLOWS_DIR)
```

```python
TARGET_INDICATORS = [
    'PA1a', 'PA1b', 'PA2a', 'PA2b', 'PA3',  'PA9',  'PA10',
    'TM1a', 'TM1b', 'TM2a', 'TM2b', 'TM5a', 'TM5b', 'TM6a', 'TM6b', 'TM7', 'TM8', 'TM9',
    'ID1a', 'ID1b', 'ID2a', 'ID2b', 'ID5a', 'ID5b', 'ID6a', 'ID6b', 'ID7', 'ID8', 'ID9',
]
```

## Load staged WIPO IP indicator data

```python
# Map indicator abbreviations to dataframes using filepaths

dfs: dict[str, pd.DataFrame] = {}  # schema: {indicator_abbr (str): indicator_data (pd.DataFrame)}
for filepath in STG_IP_INDICATORS_DIR.iterdir():
    filename_stem = filepath.stem  # <Path('.../stg_wipo__ID1a.parquet')> -> 'stg_wipo_ID1a'
    indicator_abbr = filename_stem.split('__')[1]  # 'stg_wipo__ID1a' -> 'ID1a'
    if indicator_abbr in TARGET_INDICATORS:
        print(f'Loading {filepath.name:.<22}... ', end='')
        dfs[indicator_abbr] = pd.read_parquet(filepath)
        print('Done')

print('---------------------------------')
try:
    all(indicator in TARGET_INDICATORS for indicator in dfs.keys())
    print('SUCCESS: All IP indicators loaded')
except FileNotFoundError as e:
    print(f'Not all indicators in `TARGET_INDICATORS` loaded: {e}')
```

```python
for ip_indicator in dfs:
    df = dfs[ip_indicator]
    cols_and_dtypes = {col: df[col].dtype for col in df.columns}
    print(f'{ip_indicator} columns and dtypes:')
    for text_col in cols_and_dtypes:
        print(f'    {text_col:.<27}: {df[text_col].dtype}')
    print()
```

## Incorporate Chinese TM app data into indicators TM1a and TM1b

```python
china_tm_df = (
    pd.read_csv(STG_CHINA_DIR / 'stg_china__tm_filings.csv')
      .rename(columns={'region': 'origin'})
      .drop_duplicates()
      .convert_dtypes()
)
china_tm_df['destination_office'] = 'China'
china_tm_df['destination_office_st3_code'] = 'CN'
```

```python
china_tm_df.head()
```

```python
def insert_chinese_tm_data(
    ip_indicator: str, china_tm_df: pd.DataFrame) -> pd.DataFrame:
    
    if ip_indicator == 'TM1a':
        app_col = 'direct_apps'
    elif ip_indicator == 'TM1b':
        app_col = 'madrid_apps'
    else:
        raise ValueError("Error: Incorrect indicator. "
                         "Indicator must be in ['TM1a', 'TM1b'].")

    china_cols = [
        'year',
        'origin',
        'destination_office',
        'destination_office_st3_code',
        f'{app_col}',
    ]
    china_df = china_tm_df[china_cols].copy()

    tm_df = dfs[ip_indicator].copy()
    tm_df = tm_df.merge(
        china_df,
        on=['year', 'origin', 'destination_office', 'destination_office_st3_code'],
        how='left'
    )

    # Sum while keeping NA values
    tm_total = tm_df[['count', app_col]].sum(axis=1, min_count=1)

    # Keep NA values unless:
    #   1. Original count != NA, or
    #   2. Combined count > 0 (i.e. China had data)
    tm_df['count'] = (
        tm_total
        .where(tm_df['count'].notna() | (tm_total > 0))
        .astype('Int64')
    )
    tm_df = (
        tm_df
        .drop(columns=app_col)
        .drop_duplicates()
    )

    col_order = [
        'year',
        'origin',
        'destination_office',
        'destination_office_st3_code',
        'count',
    ]
    return tm_df[col_order]
        
```

```python
dfs['TM1a']
```

```python
dfs['TM1a'] = insert_chinese_tm_data('TM1a', china_tm_df)
dfs['TM1b'] = insert_chinese_tm_data('TM1b', china_tm_df)
```

```python
dfs['TM1a']
```

### Transform IP indicator tables: Replace 'Total' with 'World'

To maintain consistency with UN M49 region naming system

```python
# Replace 'Total' -> 'World' to maintain consistency with UN M49 regions
for indicator in dfs:
    df = dfs[indicator]
    df[['origin', 'destination_office']] = (
        df[['origin', 'destination_office']]
        .replace({'Total': 'World'})
    )
```

### Transform IP indicator tables: Add ST.3 code columns

To add: `origin_st3_code` or `destination_office_st3_code`

```python
st3_df = pd.read_csv(STG_IP_OFFICES_DIR / 'stg_wipo__st3_codes.csv')
```

```python
st3_df.head()
```

```python
st3_code_map = dict(zip(st3_df['entity_name'], st3_df['st3_code']))
for indicator, df in dfs.items():
    changed = False

    if 'origin_st3_code' not in df.columns:
        print(f"{indicator:<4}: Adding col 'origin_st3_code'... ", end='')
        origin_st3_code = pd.Series(df['origin'].map(st3_code_map),
                                    dtype='string')
        df = df.assign(origin_st3_code=origin_st3_code)
        changed = True
        print('Done')

    if 'destination_office_st3_code' not in df.columns:
        print(f"{indicator:<4}: Adding col 'destination_office_st3_code'... ", end='')
        destination_office_st3_code = pd.Series(df['destination_office'].map(st3_code_map),
                                                dtype='string')
        df = df.assign(destination_office_st3_code=destination_office_st3_code)
        changed = True
        print('Done')

    if changed:
        dfs[indicator] = df
```

```python
dfs['TM9']
```

```python
# 1. Concatenate
long_df = (
    pd.concat(dfs, names=['indicator'])        # first index level = indicator key
      .reset_index(level=0)                    # bring it out to a column
      .rename(columns={'level_0': 'indicator'})
)

# 2. Cast column dtypes
text_cols = [
    'origin',
    'origin_st3_code',
    'destination_office',
    'destination_office_st3_code',
    'indicator',
]
long_df[text_cols] = long_df[text_cols].apply(lambda s: s.astype('string').str.strip())

# Year and count with proper nullable dtypes
long_df['year']  = pd.to_numeric(long_df['year'],  errors='coerce').astype('Int32')
long_df['count'] = pd.to_numeric(long_df['count'], errors='coerce').astype('Float64')

# 3. Fill 'World' rows with 'WD' ST3 value (only for World totals with
#    missing codes; some already have it)
origin_world_mask = (long_df['origin'].eq('World')
                     & long_df['origin_st3_code'].isna())
dest_world_mask   = (long_df['destination_office'].eq('World') 
                     & long_df['destination_office_st3_code'].isna())
long_df.loc[origin_world_mask, 'origin_st3_code'] = 'WD'
long_df.loc[dest_world_mask,   'destination_office_st3_code'] = 'WD'

# 4. Order columns
col_order = [
    'year',
    'origin',
    'origin_st3_code',
    'destination_office',
    'destination_office_st3_code',
    'indicator',
    'count',
]
long_df = long_df[col_order].reset_index(drop=True)
```
