## Setup

```python
from time import sleep
from pathlib import Path 
import pandas as pd 
import sys

project_root = Path.cwd().parent.parent
sys.path.insert(0, str(project_root))

from utils.paths import (STG_WORLD_BANK_DIR, STG_WORLD_BANK_DIR, STG_GAC_DIR,
                         INT_UOM_DIR, INT_MACROECON_STATS_DIR)
```

```python
stg_wb_macroecon_filename = 'stg_world_bank__macroecon_stats.csv'

macroecon_df = pd.read_csv(STG_WORLD_BANK_DIR / stg_wb_macroecon_filename)
```

## Add units of measure based on macroecon stat series

### Note: Units of measure currently unused

```python
units_of_measure_filename = 'int_units_of_measure.csv'  # TODO: Units_of_measure currently unused

uom_df = pd.read_csv(INT_UOM_DIR / units_of_measure_filename)
```

```python
macroecon_df = macroecon_df.copy()

# -----------------------------------------------------------
# 1. Extract unit of measure information from 'series' column
# -----------------------------------------------------------

# Create a temporary column 'unit_text' with content from parentheses, e.g., "constant 2017 international $"
macroecon_df['unit_text'] = macroecon_df['series'].str.extract(r'\(([^)]+)\)$', expand=False)

# Clean the 'series' column by removing the unit information in parentheses
macroecon_df['series'] = macroecon_df['series'].str.replace(r'\s*\([^)]*\)$', '', regex=True).str.strip()
```

```python
# ----------------------------------------
# Standardize unit of measure descriptions
# ----------------------------------------

def classify_unit_text(txt):
    if pd.isna(txt):
        return pd.NA
    txt_lower = txt.lower()
    if txt_lower == '$' or 'dollar' in txt_lower:
        return 'Dollars'
    if txt_lower == 'number':
        return 'Number'
    return txt.strip()

# Create 'uom' column
macroecon_df['uom'] = macroecon_df['unit_text'].apply(classify_unit_text)

# Handle the special case for population, which has no unit in its series name
is_population = macroecon_df['series'].str.contains('Population', case=False, na=False)
macroecon_df.loc[is_population, 'uom'] = 'persons'
```

```python
# Update main units_of_measure (uom) dataframe
uom_df['unit_description'] = uom_df['unit_description'].str.lower()

existing_units = set(uom_df['unit_description'])
new_units = set(macroecon_df['uom'].dropna().str.lower()) - existing_units

if new_units:
    print(f"Found {len(new_units)} new unit(s) to add: {new_units}")
    next_id = uom_df['uom_id'].max() + 1
    new_uom_rows = []
    for unit in sorted(list(new_units)): # Sorting for deterministic behavior
        new_uom_rows.append({'uom_id': next_id, 'unit_description': unit})
        next_id += 1
    
    # Concatenate new rows to the uom_df
    uom_df = pd.concat([
        uom_df,
        pd.DataFrame(new_uom_rows)
    ], ignore_index=True)
    print("UOM table updated.")
else:
    print("No new units found. UOM table is up-to-date.")
```

```python
# -------------------------------------------
# Map UOM IDs back to the Units of Measure df
# -------------------------------------------

# Create a dictionary for mapping: {'persons': 4, 'dollars': 1, ...}
uom_map = dict(zip(uom_df['unit_description'], uom_df['uom_id']))

# Map the IDs using the standardized 'uom' column
macroecon_df['uom_id'] = macroecon_df['uom'].str.lower().map(uom_map)
```

```python
# -----------------------------
# Final unit of measure cleanup
# -----------------------------

# Drop the temporary 'unit_text' column
macroecon_df = macroecon_df.drop(columns=['unit_text'])

col_order = ['wb_region_code', 'region', 'series', 'year', 'value', 'uom', 'uom_id']
macroecon_df = macroecon_df[col_order].convert_dtypes()
```

## Add region codes to macroecon df

```python
stg_gac_regions_filename = 'stg_gac__regions.csv'
stg_wb_regions_filename  = 'stg_world_bank__regions.csv'

wb_regions_df  = pd.read_csv(STG_WORLD_BANK_DIR / stg_wb_regions_filename)
gac_regions_df = pd.read_csv(STG_GAC_DIR / stg_gac_regions_filename)
```

```python
# If necessary, rename prototype ISO Alpha-3 columns to most up to date col names

wb_regions_df  =  wb_regions_df.rename(columns={'iso_alpha-3_code': 'iso_alpha_3_code'})
gac_regions_df = gac_regions_df.rename(columns={'iso_alpha-3_code': 'iso_alpha_3_code'})
```

```python
macroecon_regions_df = (
    macroecon_df
    .merge(
        gac_regions_df[['iso_alpha_3_code', 'un_m49_code']],
        left_on='wb_region_code',
        right_on='iso_alpha_3_code',
        how='left'
        )
    .merge(
        wb_regions_df[['world_bank_region_id', 'iso_alpha_3_code']],
        left_on='wb_region_code',
        right_on='iso_alpha_3_code',
        how='left'
        )
    .rename(columns={'iso_alpha_3_code_x': 'iso_alpha_3_code',
                     'world_bank_region_id': 'wb_region_id'})
    .drop(columns=['iso_alpha_3_code_y'])
    # Select columns to keep, in order
    [['year', 'region',
      'wb_region_code', 'iso_alpha_3_code', 'un_m49_code',
      'series', 'uom', 'uom_id', 'value']]
    # Replace values in `series` column with shortened lower/camelcase versions
    .replace({'Population, total': 'population',
              'GDP, PPP': 'gdp_ppp',
              'GDP per capita, PPP': 'gdp_ppp_per_capita'})
    .drop_duplicates()
)
```

```python
macroecon_regions_df
```

### Pivot macroecon df

```python
# Pivot on series so 'population', 'gdp_ppp', etc. get their own columns
id_cols = ['year', 'region', 'wb_region_code', 'iso_alpha_3_code', 'un_m49_code']

# Note: This pivot converts the cols in `id_cols` to a multi-index, so
#       they're not considered columns until reset_index() is called later
macroecon_regions_wide = (
    macroecon_regions_df
    .pivot(index=id_cols,
           columns='series',
           values=['value', 'uom', 'uom_id'])
)

macroecon_regions_wide.columns = [
    # `value_col`'s possible values are ['value', 'uom', 'uom_id']
    series if value_col == 'value' else f"{series}_{value_col}"
    for value_col, series in macroecon_regions_wide.columns
]
```

```python
# Reorder columns

wide_col_order = [
    'population', 'population_uom', 'population_uom_id',
    'gdp_ppp', 'gdp_ppp_uom', 'gdp_ppp_uom_id',
    'gdp_ppp_per_capita', 'gdp_ppp_per_capita_uom', 'gdp_ppp_per_capita_uom_id'
]
macroecon_regions_wide = macroecon_regions_wide[wide_col_order]

# Fix datatypes
macroecon_regions_wide = (
    macroecon_regions_wide
    .reset_index()
    .convert_dtypes()
    .astype({'population': 'Int64'})  # Force population to be cast as int
)
```
