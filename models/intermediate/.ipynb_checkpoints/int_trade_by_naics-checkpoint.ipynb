```python
from time import sleep
from pathlib import Path 
import pandas as pd 
import sys

project_root = Path.cwd().parent.parent
sys.path.insert(0, str(project_root))
import duckdb as dd

from utils.paths import (STG_STATCAN_DIR,
                         INT_MACROECON_FLOWS_DIR)
```

```python
trade_by_naics_filename = 'stg_statcan__trade_by_naics.parquet'
```

```python
# File & Path params
IN_TBL_NAME = 'stg_statcan__trade_by_naics'
IN_FILEPATH = STG_STATCAN_DIR / f'{IN_TBL_NAME}.parquet'

OUT_TBL_NAME = 'int_trade_by_naics'
OUT_FILEPATH = INT_MACROECON_FLOWS_DIR / f'{OUT_TBL_NAME}.parquet'

# Query parameters
industry_grp_regexp      = r'\[(\d{4})\]$'
industry_grp_name_regexp = r'\s*\[[^\]]+\]$'
```

```python
con = dd.connect()
```

```python
int_query = f"""
-- This query transforms the staged data to its final grain before ranking.
-- 1. Parses the NAICS string into a separate name and ID.
-- 2. Creates a user-friendly 'trade_direction' column.
-- 3. Handles the 'Total of all industries' special case.
-- 4. Aggregates the monthly dollar values into a yearly sum (cad_value),
--    changing the grain of the table from monthly to yearly.

WITH naics_parsed AS (
    SELECT
        year,
        trade_type,
        trading_partner,
        -- Create a descriptive trade direction
        CASE
            WHEN trade_type = 'Exports' THEN 'Canada to ' || trading_partner
            WHEN trade_type = 'Imports' THEN trading_partner || ' to Canada'
        END AS trade_direction,

        -- Keep the original NAICS string for grouping
        naics AS naics_industry_grp,

        -- Extract a clean NAICS name
        CASE
            WHEN naics = 'Total of all industries' THEN 'Total'
            ELSE TRIM(REGEXP_REPLACE(naics, '{industry_grp_name_regexp}', ''))
        END AS naics_name,

        -- Extract the numeric NAICS ID
        CASE
            WHEN REGEXP_MATCHES(naics, '{industry_grp_regexp}')
            THEN TRY_CAST(REGEXP_EXTRACT(naics, '{industry_grp_regexp}', 1) AS INT)
            ELSE NULL
        END AS naics_code,

        cad_value
    FROM
        read_parquet('{IN_FILEPATH}')
)
-- Aggregate to the yearly grain
SELECT
    year,
    trading_partner,
    trade_type,
    trade_direction,
    naics_industry_grp,
    naics_name,
    naics_code,
    SUM(cad_value) AS cad_value
FROM
    naics_parsed
GROUP BY ALL
"""

int_trade_by_naics = 'int_trade_by_naics'  # More DBT-esque name: 'int_trade_by_naics_aggregated_yearly'
con.sql(f"CREATE OR REPLACE VIEW '{OUT_TBL_NAME}' AS ({int_query})")
```

```python
save_query = f"""
COPY
    (SELECT * FROM '{OUT_TBL_NAME}')
TO
    '{OUT_FILEPATH.as_posix()}'
(FORMAT PARQUET);
"""

con.execute(save_query)
con.close()
```
