Get raw World Bank macroeconomic stats via API
==============================================

Endpoint URL of World Bank API for economic indicators: 
```url
https://api.worldbank.org/v2/country/all/indicator/
```

## Setup

````python
from pathlib import Path

import pandas as pd
import wbgapi as wb

from utils.paths import RAW_WORLD_BANK_DIR
````

````python
# World Bank indicator codes

wb_indicators = {
    'SP.POP.TOTL': 'Population, total',
    'NY.GDP.MKTP.PP.KD': 'GDP, PPP (constant 2021 international $)',
    'NY.GDP.PCAP.PP.KD': 'GDP per capita, PPP (constant 2021 international $)',
}
wb_indicator_ids = list(wb_indicators.keys())
````

## (Optional) Explore World Bank API

````python
wb.series.info(id=['SP.POP.TOTL', 'NY.GDP.MKTP.PP.KD', 'NY.GDP.PCAP.PP.KD'])
````

## Extract World Bank macroecon stats via API

````python
wb_df: pd.DataFrame = wb.data.DataFrame(wb_indicator_ids, labels=True)
wb_df
````

## Load macroecon stats to raw data directory

````python
wb_filename = 'raw_world_bank__macroecon_stats.csv'
wb_filepath: Path = RAW_WORLD_BANK_DIR / wb_filename

wb_df.to_csv(wb_filepath, index=False)
````
