```python
from time import sleep
from pathlib import Path 
import pandas as pd 
import sys

project_root = Path.cwd().parent.parent.parent
sys.path.insert(0, str(project_root))
```

```python
from utils.etl_utils import (
    get_full_table_zip_url,
    download_zip,
    extract_zip_csv_to_parquet
)
from utils.paths import RAW_STATCAN_DIR
```

```python
TRADE_BY_NAICS_TABLE_ID = 12100176  # CSV is 20.8 GB uncompressed; use parquet

tables_to_parquet = [
    TRADE_BY_NAICS_TABLE_ID,
]
```

```python
# NOTE: This takes ~15 minutes to run. 5 to download, 10 to convert to
#       parquet with current `extract_zip_csv_to_parquet()` implementation.
#       - Consider using DuckDB with built-in httpfs extension; that'd
#         be way faster.

for table_id in tables_to_parquet:
    zip_filepath     = RAW_STATCAN_DIR / f"raw_statcan__tbl_{table_id}.zip"
    parquet_filepath = RAW_STATCAN_DIR / f"raw_statcan__tbl_{table_id}.parquet"

    zip_url = get_full_table_zip_url(table_id=table_id, verbose=True)

    
    download_zip(
        url=zip_url,
        destination=zip_filepath,
        verbose=True,
    )

    extract_zip_csv_to_parquet(
        zip_path=zip_filepath,
        out_parquet=parquet_filepath,
        csv_member=f"{table_id}.csv",
    )
```
